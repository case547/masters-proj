```python
import os

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
# Values from baseline_results.csv
BASELINE_TYRE_PM = 68208.8911131483
BASELINE_FINAL_WAIT = 2722.0
```

```python
df = pd.read_csv("alpha.csv")  # cleaned results from 2way_single_intersection_results.csv

alpha = df["alpha"] 
tyre_pm = df["tyre_pm_emitted"] 
final_wait = df["final_waiting_time"]

plt.figure(figsize=(7.5,5))
plt.scatter(alpha, tyre_pm, marker=".", label="NEE Penalty")
plt.scatter(alpha, final_wait, marker=".", label="Congestion Penalty")
plt.plot(alpha, tyre_pm, "C0", linewidth=1)
plt.plot(alpha, final_wait, "C1", linewidth=1)

xmin, xmax = plt.xlim()
plt.hlines(BASELINE_TYRE_PM, xmin, xmax, "C2", "dashed", "Baseline NEE penalty")
plt.hlines(BASELINE_FINAL_WAIT, xmin, xmax, "C3", "dashed", "Baseline congestion penalty")

plt.grid()
plt.title("$\\alpha$ versus Penalty, with Baseline Comparison")
plt.xlabel("$\\alpha$")
plt.ylabel("Penalty")

plt.legend()
plt.yscale("log")
plt.show()
```
